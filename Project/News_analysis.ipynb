```python
def get_stratified_sample(df, samples_per_year=1500):
    sampled_data = []
    for year in range(2019, 2022):  # Focus on recent 3 years
        year_data = df[df['year'] == year]
        sampled_data.append(year_data.sample(n=samples_per_year, random_state=42))

    stratified_df = pd.concat(sampled_data)
    return stratified_df

# Get stratified sample
sampled_df = get_stratified_sample(df)

print(f"Total samples: {len(sampled_df)}")
print("\nSamples per year:")
print(sampled_df.groupby('year').size())
```

```python
import torch
from transformers import AutoTokenizer, AutoModel
import numpy as np
from datetime import datetime
from tqdm import tqdm
import pandas as pd

class EmbeddingGenerator:
    def __init__(self, auth_token):
        self.tokenizer = AutoTokenizer.from_pretrained(
            "meta-llama/Llama-3.2-1B",
            token=auth_token
        )
        self.tokenizer.pad_token = self.tokenizer.eos_token

        self.model = AutoModel.from_pretrained(
            "meta-llama/Llama-3.2-1B",
            token=auth_token
        )
        self.model.eval()

    def get_text_embedding(self, text, max_length=512):
        """Generate embeddings for input text"""
        inputs = self.tokenizer(
            text,
            padding=True,
            truncation=True,
            max_length=max_length,
            return_tensors="pt"
        )

        with torch.no_grad():
            outputs = self.model(**inputs)
            embeddings = outputs.last_hidden_state.mean(dim=1)
        return embeddings[0].numpy()

    def process_date(self, date_input):
        """Process date input into feature array"""
        if isinstance(date_input, str):
            date = datetime.strptime(date_input, '%Y-%m-%d')
        else:
            date = datetime.combine(date_input, datetime.min.time()) if hasattr(date_input, 'year') else date_input

        return np.array([
            date.year,
            date.month,
            date.day,
            date.weekday()
        ])

    def validate_data(self, df):
        """Validate data before processing"""
        required_columns = ['title', 'content', 'date', 'stock']
        missing_columns = [col for col in required_columns if col not in df.columns]
        if missing_columns:
            raise ValueError(f"Missing required columns: {missing_columns}")

        # Test date processing with first row
        try:
            _ = self.process_date(df['date'].iloc[0])
        except Exception as e:
            raise ValueError(f"Date processing failed on first row: {e}")

        # Check for price columns
        price_cols = [col for col in df.columns if col.startswith('ts_')]
        if not price_cols:
            raise ValueError("No price columns (ts_*) found in DataFrame")

        print("Data validation passed!")
        return True

    def test_embedding_pipeline(self, df, sample_size=2):
        """Test the full pipeline with a small sample"""
        print("Testing embedding pipeline with small sample...")
        test_df = df.head(sample_size)

        try:
            # Test text embedding
            combined_text = test_df['title'].iloc[0] + " " + test_df['content'].iloc[0]
            test_embedding = self.get_text_embedding(combined_text)
            print(f"✓ Text embedding shape: {test_embedding.shape}")

            # Test date processing
            test_date = self.process_date(test_df['date'].iloc[0])
            print(f"✓ Date feature shape: {test_date.shape}")

            # Test full pipeline
            embeddings, dims = self.create_comprehensive_embedding(test_df)
            print("✓ Full pipeline test successful!")
            print(f"✓ Feature dimensions: {dims}")

            return True
        except Exception as e:
            print(f"Pipeline test failed: {e}")
            raise

    def create_comprehensive_embedding(self, df, batch_size=32):
        """Create embeddings with progress tracking"""
        # Validate data first
        self.validate_data(df)

        # Initialize progress tracking
        total_steps = 4  # text, date, price, stock
        current_step = 0

        print(f"\nStep {current_step + 1}/{total_steps}: Generating text embeddings...")
        text_embeddings = []
        combined_texts = df['title'] + " " + df['content']

        for i in tqdm(range(0, len(df), batch_size)):
            batch_texts = combined_texts[i:i + batch_size]
            batch_embeddings = [self.get_text_embedding(text) for text in batch_texts]
            text_embeddings.extend(batch_embeddings)

        text_embeddings = np.array(text_embeddings)
        current_step += 1

        print(f"\nStep {current_step + 1}/{total_steps}: Processing date features...")
        date_features = np.array([self.process_date(date) for date in df['date']])
        current_step += 1

        print(f"\nStep {current_step + 1}/{total_steps}: Processing price features...")
        price_cols = [col for col in df.columns if col.startswith('ts_')]
        price_features = df[price_cols].values
        current_step += 1

        print(f"\nStep {current_step + 1}/{total_steps}: Processing stock features...")
        stock_encoder = pd.get_dummies(df['stock'])
        stock_features = stock_encoder.values

        # Combine all features
        combined_embeddings = np.concatenate([
            text_embeddings,
            date_features,
            price_features,
            stock_features
        ], axis=1)

        feature_dims = {
            'text': text_embeddings.shape[1],
            'date': date_features.shape[1],
            'price': price_features.shape[1],
            'stock': stock_features.shape[1]
        }

        return combined_embeddings, feature_dims

```

```python
def process_with_validation(df, auth_token):
    """Helper function to process data with validation"""
    embedder = EmbeddingGenerator(auth_token)

    # First test with a small sample
    print("\nTesting pipeline with small sample...")
    embedder.test_embedding_pipeline(df.head(2))

    # If test passes, process full dataset
    print("\nProcessing full dataset...")
    return embedder.create_comprehensive_embedding(df)
```

```python
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

def analyze_embedding_importance(embeddings, current_prices, future_prices, feature_dims, n_important=20):
    """
    Enhanced Ridge regression analysis with detailed metrics and visualizations.
    """
    # Calculate returns
    returns = (future_prices - current_prices) / current_prices * 100

    # Split data for validation
    X = StandardScaler().fit_transform(embeddings)
    X_train, X_test, y_train, y_test = train_test_split(X, returns, test_size=0.2, random_state=42)

    # Fit model
    model = Ridge(alpha=1.0)
    model.fit(X_train, y_train)

    # Make predictions
    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)

    # Calculate metrics
    metrics = {
        'Train R²': r2_score(y_train, y_pred_train),
        'Test R²': r2_score(y_test, y_pred_test),
        'Train MAE': mean_absolute_error(y_train, y_pred_train),
        'Test MAE': mean_absolute_error(y_test, y_pred_test),
        'Train RMSE': np.sqrt(mean_squared_error(y_train, y_pred_train)),
        'Test RMSE': np.sqrt(mean_squared_error(y_test, y_pred_test))
    }

    # Feature importance analysis
    importance = np.abs(model.coef_)
    feature_mapping = []
    for feature, dim in feature_dims.items():
        feature_mapping.extend([feature] * dim)

    importance_df = pd.DataFrame({
        'Feature_Type': feature_mapping,
        'Dimension': range(len(importance)),
        'Importance': importance,
        'Coefficient': model.coef_
    })

    # Aggregate by feature type
    feature_importance = importance_df.groupby('Feature_Type')['Importance'].sum()

    # Create multiple plots
    fig = plt.figure(figsize=(15, 12))

    # 1. Feature Type Importance
    plt.subplot(2, 2, 1)
    feature_importance.plot(kind='bar')
    plt.title('Feature Type Importance')
    plt.ylabel('Importance')

    # 2. Prediction vs Actual (Test Set)
    plt.subplot(2, 2, 2)
    plt.scatter(y_test, y_pred_test, alpha=0.5)
    plt.plot([returns.min(), returns.max()], [returns.min(), returns.max()], 'r--', lw=2)
    plt.xlabel('Actual Returns (%)')
    plt.ylabel('Predicted Returns (%)')
    plt.title('Prediction vs Actual (Test Set)')

    # 3. Coefficient Distribution
    plt.subplot(2, 2, 3)
    sns.histplot(model.coef_, bins=50)
    plt.title('Distribution of Coefficients')
    plt.xlabel('Coefficient Value')

    # 4. Top Individual Features
    plt.subplot(2, 2, 4)
    top_features = importance_df.nlargest(10, 'Importance')
    sns.barplot(data=top_features, x='Importance', y='Feature_Type')
    plt.title('Top 10 Individual Features')

    plt.tight_layout()
    plt.show()

    # Print detailed metrics
    print("\nModel Performance Metrics:")
    for metric, value in metrics.items():
        print(f"{metric}: {value:.4f}")

    print("\nOverall Feature Type Importance:")
    print(feature_importance.sort_values(ascending=False))

    # Additional analysis: Feature importance percentages
    print("\nFeature Type Importance (Percentage):")
    feature_importance_pct = (feature_importance / feature_importance.sum() * 100)
    print(feature_importance_pct.sort_values(ascending=False))

    # Return type contribution analysis
    print("\nReturn Analysis:")
    print(f"Mean Return: {np.mean(returns):.2f}%")
    print(f"Std Return: {np.std(returns):.2f}%")
    print(f"Min Return: {np.min(returns):.2f}%")
    print(f"Max Return: {np.max(returns):.2f}%")

    return importance_df, model, metrics
```
